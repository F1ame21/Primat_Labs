```python
# обладает линейной скоростью сходимости
def f(x):
  y =  np.sin(x) - np.log(x * x) + 10 
  return y
print("Left border:")
a = float(input())
print("Right border:")
b = float(input())
print("ε:")
e = float(input())
# количество интераций
iter_amount = 0

golden_ratio_value_1 = (3 - np.sqrt(5)) / 2
golden_ratio_value_2 = 1 - golden_ratio_value_1

x_1 = a + golden_ratio_value_1 * (b - a)
x_2 = a + golden_ratio_value_2 * (b - a)

x_1_value = f(x_1)
x_2_value = f(x_2)
# уменьшаем на окнстанту золотого сечения (3 - np.sqrt(5)) / 2
while (b - a) / 2 > e:
  if x_1_value < x_2_value:
    b = x_2
    x_2 = x_1
    x_2_value = x_1_value
    x_1 = a + golden_ratio_value_1 * (b - a)
    x_1_value = f(x_1)
  elif x_1_value > x_2_value:
    a = x_1

    x_1 = x_2
    x_1_value = x_2_value
    x_2 = a + golden_ratio_value_2 * (b - a)
    x_2_value = f(x_2)
  else:
    a = x_1
    b = x_2
    x_1 = a + golden_ratio_value_1 * (b - a)
    x_2 = a + golden_ratio_value_2 * (b - a)
    x_1_value = f(x_1)
    x_2_value = f(x_2)
  iter_amount += 1
  print(f"Left border: {a}")
  print(f"Right border: {b}\n")
  minimum = (a + b) / 2
print(f"Iter amount: {iter_amount}")
print(f"Minimum: {minimum}")
print(f"Minimum value: {f(minimum)}")
```
