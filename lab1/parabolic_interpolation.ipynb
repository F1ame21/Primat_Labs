```python
# метод с суперлинейной скоростью сходимости, однако такая высокая скорость сходимости гарантируется только в малой окрестности точки минимума
# на начальных он может делать либо слишком большие шаги, либо слишком маленькие которые будут приводить к неустойчивому поведению, 100% сходимости нет
def f(x):
  y =  np.sin(x) - np.log(x * x) + 10 
  return y
print("Left border:")
a = float(input())
print("Right border:")
b = float(input())
print("Middle point:")
middle_point = float(input())  # 4
print("ε:")
e = float(input())
delta = e / 2
# счетчик итераций
iter_amount = 0
def print_params(a, middle_point, b):
    print("Left border: ", a, end=", ")
    print("Middle point: ", middle_point, end=", ")
    print("Right border: ", b, end=".\n")

def condition(f, a, b, middle_point):
  if (f(middle_point) < f(a)) and (f(middle_point) < f(b)):
    return True
  return False
# точка минимума нашей параболы 
def calculate(y, x_1, x_2, x_3):
    f_1 = f(x_1)
    f_2 = f(x_2)
    f_3 = f(x_3)
    numerator = (x_2 - x_1) ** 2 * (f_2 - f_3) - (x_2 - x_3) ** 2 * (f_2 - f_1)
    denominator = (x_2 - x_1) * (f_2 - f_3) - (x_2 - x_3) * (f_2 - f_1)
    u = x_2 - numerator / (2 * denominator)
    return u 

while condition(f, a, b, middle_point) == False:
  print("Choose any other middle point")
  middle_point = float(input())
flag = True
while flag:
  print_params(a, middle_point, b)
  iter_amount +=1
  if ((b - a) <= e) or ((b - middle_point) <= e) or ((middle_point) - a <= e):
    print("stop")
    flag = False
  x_1 = a
  x_2 = middle_point
  x_3 = b

  u = calculate(f, x_1, x_2, x_3)
  min = u

  if x_1 < u < x_2:
    a = x_1
    middle_point = u
    b = x_2
  elif x_2 < u < x_3:
    a = x_2
    middle_point = u
    b = x_3
  elif x_1 < u < x_3:
    a = x_1
    middle_point = u
    b = x_3
  else:
    print_params(a, middle_point, b)
    flag = False
print(f"Min(x): x = {min}, Min(y): y = {f(min)}")
print(f"Number of iterations - {iter_amount}")
```
